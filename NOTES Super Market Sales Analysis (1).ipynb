# EDA Preparation

What was the best Product line for sales? How much was?

```python
sales_Product_line=df.groupby(['Product line']).sum()
sales_Product_line.rename(columns={'Rating':"Sum of Rating"},inplace=True)
sales_Product_line
```

#best product line is Food and beverages,total sale = 56144.8440

#What was the best Branch for sales? How much was?

```python
Branch_sales=df.groupby(['Branch']).sum() 
Branch_sales
```

the highest sales for Branch  is 'C' and sales are = 110568.7065

# To finds payment methods

```python
sales_Payment=df.groupby(['Payment']).sum()
sales_Payment.head()
```

```python
#adding month column
df['Month'] = df['Date'].str[0:1]
df.head()
```

```python
#to find sales by months
sales_Month=df.groupby(['Month']).sum()
sales_Month
```

```python
Month_Branch = df.groupby(['Month','Branch']).sum()
Month_Branch
```

```python
Month_Branch_Product_line= df.groupby(['Month','Branch','Product line']).sum()
Month_Branch_Product_line
```

```python
#to sales of gender
sales_Branch=df.groupby(['Gender']).sum()
sales_Branch.head()
```

#high sales by Females

```python
#To find avg rating of product line
Product_line= df[['Product line']]
Rating = df[['Rating']]
df_final = pd.concat([Product_line,Rating],axis=1)
print(df_final.head())
print(df_final.shape)
```

```python
Avg_Rating = df_final.groupby(['Product line']).mean()
Avg_Rating
```

```python
df_final2= pd.concat([sales_Product_line,Avg_Rating],axis=1)
df_final2
```

the highest avg rating of product line is Food and beverages = 7.113218
